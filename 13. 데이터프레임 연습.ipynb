```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family']='Malgun Gothic'
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
# 미국에서 태어난 남아 여야 수

birthDF = pd.read_csv('data/births.txt', header=None, index_col=0)
birthDF.columns = ['boys', 'girls']
birthDF.index.name = 'year'
birthDF
```

```python
#1.남아수 총합을 구하시요

birthDF["boys"].sum()
birthDF.sum()['boys']
```

```python
#2.여아수 평균을 구하시요

birthDF['girls'].mean()
birthDF.mean()['girls']
```

```python
#3. 남아수가 가장많은 년도와 남아수를 구하시요
#print(birthDF['boys'].max())
#print(birthDF['boys'].idxmax())
print(birthDF['boys'].nlargest(1,keep='all'))

m = birthDF['boys'].max()
birthDF.query(f'boys=={m}')['boys']
```

```python
#4.2000년도 이후 데이터를 구하시요

birthDF.loc[2000:]
birthDF.query('year>=2000')
```

```python
#5. 1995~ 2000 년도 데이터를 구하시요

birthDF.loc[1995:2000]
birthDF.query('1995<=year<=2000')
```

```python
#6. 남아출생이 가장많은 top5 를 구하시요(년도와 남아수)

birthDF['boys'].nlargest(5, keep="all")
birthDF.nlargest(5, 'boys',keep='all')[['boys']]
```

```python
#7. 남아율(%) 컬럼을 추가하고 각년도별 남아수가 차지하는비율을 데이터로 보여주시요.
birthDF['boys']/birthDF.sum(axis=1)
birthDF['rate(%)'] = ((birthDF['boys']/(birthDF['boys']+birthDF['girls'])*100)).round(3)

birthDF
```

```python
#8. 2000년도 이후데이터 에 대해 출생량 컬럼을 추가하고 남아수가 1800000 이상이면 '많음'아니면 '적음' 이라고 보여 주시요
#남아수 출생량
#2000   1233   '많음'
#2001   1234   '적음'

birthDF['출생량'] = birthDF['boys'].apply(lambda v : '많음' if v >= 1800000 else '적음')
birthDF.loc[2000:]
```

```python
#9.1800년대,1900년대,2000년대 각 남아,여야수 평균을 구하시요
data = [birthDF.loc[1800:1899].mean(),
        birthDF.loc[1900:1999].mean(),
        birthDF.loc[2000:2099].mean()]
myDF = pd.DataFrame(data, index=['1800','1900','2000'])
myDF.round(2).T
```
