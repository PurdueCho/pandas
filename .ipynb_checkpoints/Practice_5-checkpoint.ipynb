```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
from datetime import datetime 
from pandas_datareader import data
import seaborn as sns

matplotlib.rcParams['font.family']='Malgun Gothic'
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
titanic =sns.load_dataset('titanic')
titanic
```

```python
# 1.deck컴럼을삭제하고나이(age)의nan데이터를 나이평균값으로채우시요. 
titanic.drop(columns='deck', inplace=True)
titanic['age'].fillna(titanic['age'].mean(), inplace=True)
```

```python
titanic
```

```python
#2. 생존자와 사망자 각각의 수를 구하시요 
titanic['survived'].value_counts()
```

```python
#3.등급별(pclass) 생존평균값을 구하시요
g = titanic.groupby( 'pclass' )
data_g = g.sum()['survived']/g.count()['survived']
data_g
```

```python
#4.혼자탑승하지 않은 사람과 혼자탑승한 사람의 생존평균값을구하시요
g2 = titanic.groupby('alone')
g2.sum()['survived']/g2.count()['survived']
```

```python
#5.성별 생존한 사람의 평균값을 구하시요
g3 = titanic.groupby('sex')
data_g3 = g3.sum()['survived']/g3.count()['survived']
data_g3
```

```python
#6.등급별 티켓비용(fare)의 평균을 바차트로그리시요
data = g.mean()['fare']
data.plot(kind='bar')
plt.show()
```

```python
#7.나이분류컬럼을추가하시요
#1~15(미성년자),15~25(청년),25~35(중년),
#35~60(장년),60~(노년)으로표시하시요.
def category_age( age ):
    if  age <= 15:
        return '미성년자'
    elif 15 < age <= 25:
        return '청년'
    elif 25 < age <= 35:
        return '중년'
    elif 35 < age <= 60:
        return '장년'
    else:
        return '노년'

titanic['나이분류'] = titanic['age'].apply(category_age)
titanic
```

```python
# 8.나이분류를파이차트로표시하시요
data = titanic['나이분류'].value_counts()
data.plot(kind='pie')
plt.show()
```

```python
#9. 나이분류별 생존한 사람의 평균중 생존이가장 높은 나이분류를구하시요
g4 = titanic.groupby('나이분류')
data_g4 = g4.sum()['survived']/g4.count()['survived']
data_g4.idxmax()
```

```python
#10.성별,나이분류별,등급(pclass)을 멀티인덱스로, 컬럼은 생존 평균값으로 표시하시요
df1 = pd.DataFrame(data_g3)
df2 = pd.DataFrame(data_g4)
df3 = pd.DataFrame(data_g)
myDF = pd.concat( [df1, df2, df3])

myDF
```
